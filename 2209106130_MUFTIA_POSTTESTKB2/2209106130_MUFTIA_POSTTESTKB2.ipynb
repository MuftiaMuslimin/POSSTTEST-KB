
Memvisualisasikan datset dengan grafik Bar Plot

```python
perbandingan_volume = df['Volume'].value_counts()
perbandingan_volume.plot.bar()
plt.title('PERBANDINGAN VOLUME')
plt.xlabel('VOLUME')
plt.ylabel('BANYAKNYA VOLUME')
plt.show()
```

Korelasi Dataset Menggunakan Heatmap


```python
High_Low_corr = df.loc[:, 'High':'Low'].corr()
High_Low_corr
```

```python
data = {'High': [0.057941, 0.058719, 0.058914, 0.060469], 'Low': [0.057941, 0.058331, 0.059886, 0.059497]}
df = pd.DataFrame(data)
High_Low_corr = df.corr()
sns.heatmap(data=High_Low_corr, annot=True)
plt.show()
```

1. Rata-Rata, Nilai Minimum, Standar Deviasi, Nilai Maksimum, Q1, Q2, dan Q3 dari Attribute Bertipe Angka:
   Menggunakan `df.describe()`, kita dapat menampilkan statistik deskriptif dari kolom bertipe numerik.
   Hasil ini menunjukkan ringkasan statistik penting dari dataset, seperti:
   - Mean (Rata-rata): Memberikan informasi rata-rata dari nilai-nilai di setiap kolom.
   - Min (Nilai Minimum): Nilai terendah di setiap kolom.
   - Std (Standar Deviasi): Mengukur seberapa tersebar data dari rata-rata.
   - Max (Nilai Maksimum): Nilai tertinggi di setiap kolom.
   - Q1, Q2, Q3 (Kuartil): Menunjukkan distribusi data di antara kuartil pertama, median (kuartil kedua), dan kuartil ketiga.

2. Menampilkan 10 Record Pertama dari Dataset dengan Attribute Bertipe Angka:
   Dengan menampilkan 10 baris pertama dari kolom bertipe numerik menggunakan `df.select_dtypes('number').head(10)`,
   kita bisa melihat bagaimana data awalnya tersusun dan menilai apakah terdapat pola atau tren tertentu yang mungkin
   terlihat dari beberapa baris awal.

3. Informasi Jumlah Record, Jumlah Attribute, Nama Attribute, dan Jumlah Attribute dengan Tipe Angka:
   - Jumlah Record: Dengan `len(df)`, kita mendapatkan jumlah baris dalam dataset, yaitu berapa banyak observasi yang tercatat.
   - Jumlah Atribut: Dengan `len(df.columns)`, kita mengetahui jumlah kolom atau fitur dalam dataset.
   - Jumlah Atribut dengan Tipe Data Angka: Dengan `len(df.select_dtypes(include=['int64', 'float64']).columns)`, kita tahu
   berapa banyak kolom yang memiliki tipe data numerik.

4. Menampilkan Jumlah Nilai dari Semua Attribute yang Bernilai Object:
   Fungsi `df.select_dtypes('object').value_counts()` menampilkan jumlah kemunculan setiap nilai unik dalam kolom bertipe `object`.
   Ini berguna untuk memahami distribusi data kategori atau teks.

5. Visualisasi Dataset dengan Grafik Bar Plot:
   Pada bagian ini, kita menggunakan plot batang (`bar plot`) untuk memvisualisasikan distribusi dari kolom "Volume". Ini memberikan
   gambaran visual mengenai seberapa sering setiap nilai "Volume" muncul di dalam dataset. Grafik ini membantu mengidentifikasi apakah
   ada nilai volume yang dominan atau distribusi volume yang merata.

6. Korelasi Dataset Menggunakan Heatmap:
   Heatmap yang dihasilkan dari korelasi antara kolom "High" dan "Low" menunjukkan hubungan antar variabel dalam dataset. Korelasi
   yang lebih mendekati 1 atau -1 mengindikasikan hubungan yang kuat antara variabel. Dengan visualisasi heatmap, kita dapat dengan
   mudah mengidentifikasi korelasi positif atau negatif yang signifikan.

- Statistik deskriptif dan korelasi membantu memahami karakteristik dasar data numerik serta hubungan antar variabel.

korelasi membantu menganalisis hubungan antar variabel numerik.
